```python
letters = soup.select('div.word_test_table > table > tbody > tr > td.tc')
letters = letters[2880:] # 필요한 텍스트만 뽑기
```

```python
kanji = []
for i in range(720) :
    kanji.append(letters[i*3+1].text)
len(kanji)
```

```python
hiragana = []
for i in range(720) :
    hiragana.append(letters[i*3+2].text)
for i in range(720) :   # 글자 전처리                           
    for j in range(len(hiragana[i])) :
        if hiragana[i][j] == '\u3000' :
            hiragana[i] = hiragana[i][0:j]
len(hiragana)
```

```python
koreans = soup.select('div.cont')
korean = []
for i in range(720) :
    korean.append(koreans[i].text)
len(korean)
```

```python
df_korean = pd.Series(korean)
df_hiragana = pd.Series(hiragana)
df_kanji = pd.Series(kanji)
```

```python
df['word'] = df_kanji
df['pronunciation'] = df_hiragana
df['meaning'] = df_korean
```
